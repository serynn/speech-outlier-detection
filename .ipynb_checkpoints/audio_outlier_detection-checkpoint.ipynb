## Analysis

```python
wav_loc = "./../zh-cn_001_F_10-40_Bixby_4.wav"
wav_loc = TRAIN_DIR + 'de-DE_Bahner_F_10-50_Germany _Bixby_2.wav'

sr, wav = wavfile.read(wav_loc)
wav.shape
```

```python
def audio_info(wav_loc):
#     sr, wav = wavfile.read(wav_loc)
    wav, sr = librosa.core.load(wav_loc, 
                               sr=None, 
                              )
    wav = wav / np.abs(wav).max()
    sec = np.arange(len(wav))/float(sr)
    print('Sample Rate :', sr)  #   / 1
    print('Whole data :', wav.shape)  #   (samplerate*)
    print('Data Length(sec) :', sec[-1])  #  ()
    print('Max Amplitude :', wav.max())
    print('Min Amplitude :', wav.min())

    plt.figure(figsize=(20,6))
    plt.plot(sec, wav)
    plt.tick_params( length=3, labelsize=14, labelcolor='white', top=True)
    plt.xlabel('Time(sec)', fontdict={'color':'white', 'size':20})
    plt.ylabel('Scailed Amplitude', fontdict={'color':'white', 'size':20})

    IPython.display.Audio(data=wav, rate=sr)
```

```python
wav_loc = "./../zh-cn_001_F_10-40_Bixby_4.wav"
wav_loc = TRAIN_DIR + 'de-DE_Bahner_F_10-50_Germany _Bixby_8.wav'
audio_info(wav_loc)
```

### scipy.signal find_peaks
- height: peak minimum height
- threshold: peak    ,     
- distance: peak minimum x 
- prominence: 

```python
plt.figure(figsize=(30,40))

plt.subplot(4,1,1)
peaks, _ = find_peaks(wav, distance=200, height=500)
print(len(peaks), '/', len(wav))
plt.plot(wav)
plt.plot(peaks, wav[peaks], "or", markersize=10)

plt.subplot(4,1,2)
peaks, _ = find_peaks(wav, prominence=1000, height=500)
print(len(peaks), '/', len(wav))
plt.plot(wav)
plt.plot(peaks, wav[peaks], "or", markersize=10)

plt.subplot(4,1,3)
peaks, _ = find_peaks(wav, width=80)
print(len(peaks), '/', len(wav))
plt.plot(wav)
plt.plot(peaks, wav[peaks], "or", markersize=10)
```
